```python
import pandas as pd
```

```python
data = pd.read_csv('/content/cleaned_data.csv')
data.head()
```

Application of One-Hot Encoding to categorical features


```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)
data_encoded = encoder.fit_transform(data[['city', 'cuisine']])
one_hot_df = pd.DataFrame(data_encoded, columns=encoder.get_feature_names_out(['city', 'cuisine']))
```

```python
one_hot_df.head()
```

```python
import pickle
with open("encoder.pkl", "wb") as file:
    pickle.dump(one_hot_df, file)
```

```python
data.loc[:,['cost','rating']].shape
```

```python
one_hot_df.shape
```

```python
encoded_data = pd.concat([data.loc[:,['cost','rating']], one_hot_df], axis=1)
encoded_data.head()
```

```python
encoded_data.shape
```

```python
encoded_data.to_csv('encoded_data.csv', index=False)
```
